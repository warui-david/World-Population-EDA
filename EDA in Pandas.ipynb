```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"D:\Downloads\world_population.csv")
df
```

```python
## Setting the values to float format with 2 decimal places - from scientific notation

pd.set_option('display.float_format', lambda x: '%.2f'% x)
df
```

```python
df.info()
```

```python
## Getting high level info about the data

df.describe()
```

```python
## Checking the total number of null values

df.isnull().sum()
```

```python
## Checking non unique values

df.nunique()
```

```python
## Sorting the data by specific column

df.sort_values(by='2022 Population', ascending = False).head(10)
```

```python
## Finding the correlation between the columns

df.corr()
```

```python
## Creating a basic plot of the correlation

sns.heatmap(df.corr(), annot=True)
plt.rcParams['figure.figsize'] = (20,10)
plt.show()
```

```python
## Grouping the data by column

df.groupby('Continent').mean()
```

```python
## Finding only continents whose name contains 'Oceania'

df[df['Continent'].str.contains('Oceania')]
```

```python
## Grouping the data and using continent as the index

df.groupby('Continent').mean().sort_values(by='2022 Population',ascending=False)
```

```python
## Sorting the dataframe

df2 = df.groupby('Continent')['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population'].mean().sort_values(by='2022 Population',ascending=False)
df2
```
